Plot the top 10 countries in happiness and the bottom 5 countries

```python
df_sorted = df.sort_values('Ladder score',ascending=False)
df_sorted.head()
```

```python
color = ["#f94144","#f3722c","#f8961e","#f9c74f","#90be6d","#43aa8b","#577590"]
```

```python
top_10 = df_sorted.head(10)
bottom_10 = df_sorted.tail(10)

fig,ax = plt.subplots(2,1,figsize=(10,8),sharex=True)
sns.despine(left=True,bottom=True)


sns.barplot(data=top_10,x=top_10['Ladder score'],y=top_10['Country name'],ax=ax[0],color=color[4]).set(xlabel=None)
ax[0].set_title('top 10 countries in happiness ',size=20)
#ax[0].annotate('top 10 countries in happiness ',xy=(8,5),size=20)

ax[0].spines['bottom'].set_visible=False

sns.barplot(data=bottom_10,x=bottom_10['Ladder score'],y=bottom_10['Country name'],ax=ax[1],color=color[0])
ax[1].set_title('bottom 10 countries in happiness ',size=20)

#ax[1].annotate('bottom 10 countries in happiness',(8,5),size=20);

```

happiness by region

```python
df_region = df.groupby('Regional indicator')['Ladder score'].mean()
df_region = df_region.reset_index().sort_values('Ladder score')
df_region.head()
```

```python
df[df['Country name']=='Finland']['Ladder score']
```

```python
sns.set_style('dark')

plt.figure(figsize=(10,7))
sns.kdeplot(df_region['Ladder score'],fill=True,color=color[6],edgecolor='black',linestyle='--',linewidth=3)
plt.axvline(df_region['Ladder score'].mean(),c='k',linestyle='--')
plt.annotate('mean',xy=(5.5,0.3),rotation=90)
plt.axvline(df[df['Country name']=='Finland']['Ladder score'].values,color=color[0])
plt.annotate('Finland',xy=(8,0.3),rotation=90)
plt.axvline(df[df['Country name']=='United States']['Ladder score'].values,color=color[1])

plt.axvline(df[df['Country name']=='Afghanistan']['Ladder score'].values,color=color[3])
plt.axvline(df[df['Country name']=='India']['Ladder score'].values,color=color[4])
plt.axvline(df[df['Country name']=='China']['Ladder score'].values,color=color[6])

plt.annotate('Finland',xy=(8,0.3),rotation=90,weight='bold',size=12)
plt.annotate('USA',xy=(6.7,0.3),rotation=90,weight='bold',size=12)
plt.annotate('Afghanistan',xy=(2.2,0.3),rotation=90,weight='bold',size=12)
plt.annotate('India',xy=(3.5,0.3),rotation=90,weight='bold',size=12)
plt.annotate('China',xy=(5.1,0.3),rotation=90,weight='bold',size=12)

plt.xlim(2,10)
plt.title('Ladder Score Distribution and Mean',weight='bold',size=20);
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(15,7))
sns.barplot(data=df_region,x='Regional indicator',y='Ladder score')
plt.axhline(df_region['Ladder score'].mean())
plt.annotate('mean',(0.5,5.9),weight='bold')
plt.xticks(rotation='vertical',size=15)
plt.title('Happiness score by region',weight='bold',size=20)
plt.xlabel('');
```
